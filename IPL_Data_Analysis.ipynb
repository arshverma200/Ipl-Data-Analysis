#**2.What is Toss Decision Ratio in IPL?**

```python
toss_decision_ratio = ipl.TossDecision.value_counts()
toss_decision_ratio
```

```python
labels = (np.array(toss_decision_ratio.index))
sizes = (np.array((toss_decision_ratio)))
colors = ['orange','green']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.3f%%', shadow=True, startangle=90)
plt.title("Ratio of toss Decision in IPL",fontweight="bold",fontsize=10)
plt.show()
```

# **3.Is win depend on winning toss?**

```python
toss_vs_matchwin=(ipl.TossWinner==ipl.WinningTeam).value_counts()
toss_vs_matchwin
```

```python
labels = (np.array(toss_vs_matchwin.index))
sizes = (np.array((toss_vs_matchwin)))
colors = ['brown','yellow']

plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90)
plt.title("Toss vs Match win",fontweight="bold",fontsize=15)
plt.show()
```

#### **In most of the cases, team win the match after winning toss.**

# **6.Which Team Has won the highest number of ipl tournaments?**

```python
tournament_wins = ipl['WinningTeam'].value_counts()
highest_wins_team = tournament_wins.idxmax()
matches_of_highest_wins_team = ipl[ipl['WinningTeam'] == highest_wins_team]

plt.figure(figsize=(10, 6))
sns.histplot(data=matches_of_highest_wins_team, x='Team1')
plt.title(f'Distribution of Matches Won by {highest_wins_team} Across Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Matches Won')
plt.xticks(rotation=45)
plt.show()
```

### **Thus Mumbai Indians won the highest number of Tournaments in ipl**

# **7.Which opposition has the Mumbai Indians scored most runs against?use a scatterplot to represent it?**

```python
mumbai_wins = ipl[ipl['WinningTeam'] == 'Mumbai Indians']
runs_scored_against_opposition = mumbai_wins.groupby('Team2')['WonBy'].sum().reset_index()
most_runs_opposition = runs_scored_against_opposition.loc[runs_scored_against_opposition['WonBy'].idxmax(), 'Team2']
matches_against_most_runs_opposition = ipl[(ipl['WinningTeam'] == 'Mumbai Indians') & (ipl['Team2'] == most_runs_opposition)]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=matches_against_most_runs_opposition, x='Season', y='MatchNumber', hue='Venue', palette='viridis')
plt.title(f'Scatter Plot of Matches Won by Mumbai Indians Against {most_runs_opposition}')
plt.xlabel('Season')
plt.ylabel('Match Number')
plt.legend(title='Venue',  bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.show()
```

It implies that mumbai indians had won most matches against chennai super kings.

#**9.Which Team have Won the most number of tosses?**

```python
toss_wins = ipl['TossWinner'].value_counts()
most_toss_wins_team = toss_wins.idxmax()
matches_won_toss = ipl[ipl['TossWinner'] == most_toss_wins_team]

plt.figure(figsize=(10, 6))
sns.histplot(data=matches_won_toss, x='Season',color="orange")
plt.title(f'Distribution of Toss Wins for {most_toss_wins_team}')
plt.xlabel('Season')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

It shows mumbai has won the most number of tosses
